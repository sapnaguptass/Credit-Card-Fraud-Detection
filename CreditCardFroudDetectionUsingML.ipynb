# Data Cleansing

We have identified the input features and the target variable so we will separate them into two objects ‘X’ and ‘y’ and draw the histogram of all the input features to see the data at a glance.

```python
# V1 to V28 + Amount are features and lebel is class
features = data.iloc[:,1:30]
```

```python
features.head()
```

```python
target = data.iloc[:,30:]
target.head()
```

```python
X = features
Y = target
```

## Plot histogram of each parameter

```python
X.hist(figsize = (28,28))
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaler.fit(X)

# Plot histograms of each parameter 
X.hist(figsize = (28, 28))
plt.show()
```

# Get the training and testing set

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y)
```

# Supervised Machine Learning Model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
# Train the model using 'fit' method
model.fit(xtrain,ytrain)
```

```python
# Test the model using 'predict' method
ypred = model.predict(xtest)
```

```python
model.score(X,Y)
```
